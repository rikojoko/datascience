## Simulasi Sederhana NLP: Part-of-Speech Tagging (POS Tagging)

**Part-of-Speech Tagging (POS Tagging)** adalah proses memberi label pada setiap kata dalam teks dengan kategori gramatikalnya (misalnya, kata benda, kata kerja, kata sifat, adverbia, dll.). Ini adalah langkah penting dalam banyak aplikasi NLP karena membantu memahami struktur dan makna kalimat.

### Mengapa POS Tagging Penting?

*   **Word Sense Disambiguation:** Membedakan makna kata yang sama dengan kategori gramatikal yang berbeda (misal: "bank" sebagai institusi finansial vs. "bank" sebagai tepi sungai).
*   **Information Extraction:** Membantu mengidentifikasi jenis entitas dalam teks (misal: kata benda seringkali merupakan nama entitas).
*   **Machine Translation:** Membantu dalam penerjemahan dengan memahami peran gramatikal kata dalam kalimat.
*   **Text Analysis:** Memberikan wawasan tentang struktur bahasa yang digunakan dalam sebuah dokumen.

### Contoh Tag POS Sederhana (dalam Bahasa Inggris):

*   **NN:** Noun, singular (kata benda tunggal)
*   **NNS:** Noun, plural (kata benda jamak)
*   **VB:** Verb, base form (kata kerja bentuk dasar)
*   **VBD:** Verb, past tense (kata kerja bentuk lampau)
*   **JJ:** Adjective (kata sifat)
*   **RB:** Adverb (kata keterangan)
*   **DT:** Determiner (penentu, misal: 'the', 'a')

Untuk bahasa Indonesia, ada set tag yang berbeda, dan seringkali membutuhkan model yang dilatih khusus.

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import nltk

# Ensure the necessary NLTK resources are downloaded for tokenization and POS tagging
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True) # Added to fix the LookupError

# Teks contoh (dalam Bahasa Inggris karena tagger default NLTK berbahasa Inggris)
english_text = "The quick brown fox jumps over the lazy dog."

print(f"Teks Asli: {english_text}\n")

# 1. Tokenisasi
tokens_en = word_tokenize(english_text)
print(f"Token Bahasa Inggris: {tokens_en}\n")

# 2. POS Tagging
pos_tags_en = pos_tag(tokens_en)
print(f"POS Tags Bahasa Inggris: {pos_tags_en}")
```

### Penjelasan Simulasi POS Tagging

1.  **Pengunduhan Data NLTK:** Kode `nltk.download('averaged_perceptron_tagger')` mengunduh model yang diperlukan oleh NLTK untuk melakukan POS tagging pada teks berbahasa Inggris.
2.  **Teks Contoh:** Kita menggunakan kalimat berbahasa Inggris karena model POS tagger bawaan NLTK adalah untuk Bahasa Inggris.
3.  **Tokenisasi:** Sama seperti sebelumnya, `word_tokenize()` memecah kalimat menjadi kata-kata.
4.  **POS Tagging:** Fungsi `pos_tag()` mengambil daftar token dan mengembalikan daftar *tuple*, di mana setiap *tuple* berisi kata dan tag POS yang sesuai.

Perhatikan bahwa untuk POS tagging bahasa Indonesia, Anda akan memerlukan model dan pustaka NLTK yang dilatih khusus untuk bahasa Indonesia (misalnya `Indonesian-POS-Tagger` atau model dari `Stanza`/`SpaCy` dengan dukungan bahasa Indonesia), karena tagger bawaan NLTK kurang akurat untuk bahasa selain Inggris.

```python
import stanza

# Unduh model bahasa Indonesia
# Ini hanya perlu dilakukan sekali
try:
    stanza.download('id')
except Exception as e:
    print(f"Stanza model 'id' already downloaded or failed to download: {e}")

# Inisialisasi pipeline NLP untuk bahasa Indonesia dengan komponen POS Tagging
nlp = stanza.Pipeline(lang='id', processors='tokenize,pos')
```

Penjelasan Simulasi POS Tagging Bahasa Indonesia dengan Stanza

1.  **Instalasi Stanza:** Kita menggunakan `!pip install stanza` untuk menginstal pustaka Stanza di lingkungan Colab.
2.  **Pengunduhan Model:** `stanza.download('id')` mengunduh model NLP yang telah dilatih khusus untuk bahasa Indonesia. Ini hanya perlu dilakukan sekali.
3.  **Inisialisasi Pipeline:** `stanza.Pipeline(lang='id', processors='tokenize,pos')` membuat objek *pipeline* NLP Stanza. Kita menentukan `lang='id'` untuk bahasa Indonesia dan `processors='tokenize,pos'` untuk hanya mengaktifkan tokenisasi dan POS tagging (meskipun Stanza bisa melakukan lebih banyak seperti NER, parsing, dll.).
4.  **Pemrosesan Teks:** Teks bahasa Indonesia (`indonesian_text`) kemudian dilewatkan ke objek `nlp` untuk diproses.
5.  **Menampilkan Hasil:** Kita mengiterasi setiap kata dalam kalimat yang diproses dan mencetak kata beserta *Universal Part-of-Speech* (UPOS) tag-nya. UPOS adalah standar tag POS universal yang mempermudah perbandingan antar bahasa.

Dengan cara ini, Anda bisa melakukan POS Tagging yang lebih akurat untuk teks bahasa Indonesia, yang merupakan langkah fundamental dalam banyak tugas NLP kompleks seperti ekstraksi informasi atau analisis sintaksis.

```python
# Teks contoh dalam Bahasa Indonesia
indonesian_text = "Analisis sentimen adalah proses menentukan nada emosional di balik sepotong teks."

print(f"Teks Asli: {indonesian_text}\n")

# Proses teks menggunakan pipeline Stanza
doc = nlp(indonesian_text)

# Tampilkan hasil POS Tagging
print("POS Tags Bahasa Indonesia:")
for sentence in doc.sentences:
    for word in sentence.words:
        print(f"{word.text}: {word.upos}")
```

Setiap kata berhasil diberi label Universal Part-of-Speech (UPOS) oleh Stanza:

* Analisis: PROPN (Proper Noun - Kata Benda Nama Diri)
* sentimen: NOUN (Noun - Kata Benda)
* adalah: AUX (Auxiliary Verb - Kata Kerja Bantu)
* proses: NOUN
* menentukan: VERB (Verb - Kata Kerja)
* nada: NOUN
* emosional: ADJ (Adjective - Kata Sifat)
* di: ADP (Adposition - Kata Depan)
* balik: NOUN
* sepotong: NOUN
* teks: NOUN
.: PUNCT (Punctuation - Tanda Baca)

```python
import nltk
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)
from nltk.corpus import wordnet

# Kata yang ingin dicari sinonimnya
word_to_find_synonyms = "komputer"

print(f"Mencari sinonim untuk kata: '{word_to_find_synonyms}'\n")

synonyms = []
for syn in wordnet.synsets(word_to_find_synonyms, lang='eng'): # 'eng' untuk bahasa Inggris, WordNet utamanya berbahasa Inggris
    for lemma in syn.lemmas(lang='eng'): # Mengambil lemma dalam bahasa Inggris
        synonyms.append(lemma.name())

# Hapus duplikasi dan tampilkan
print(f"Sinonim yang ditemukan untuk '{word_to_find_synonyms}': {set(synonyms)}")

# --- Contoh untuk kata lain (misalnya dalam Bahasa Indonesia) ---
# WordNet secara native tidak mendukung Bahasa Indonesia secara langsung.
# Namun, jika ada kata Inggris yang memiliki makna sama, bisa digunakan.
# Untuk Bahasa Indonesia yang spesifik, Anda mungkin perlu kamus sinonim Bahasa Indonesia atau sumber lain.

word_to_find_synonyms_id = "cepat"
print(f"\nMencoba mencari sinonim (melalui terjemahan) untuk kata: '{word_to_find_synonyms_id}'\n")

synonyms_id = []
# Anda perlu mengetahui terjemahan yang relevan dalam bahasa Inggris
english_equivalent = "fast"
for syn in wordnet.synsets(english_equivalent, lang='eng'):
    for lemma in syn.lemmas(lang='eng'):
        synonyms_id.append(lemma.name())

print(f"Sinonim (berdasarkan '{english_equivalent}') untuk '{word_to_find_synonyms_id}': {set(synonyms_id)}")
```

Penjelasan Kode

1.  **Unduh WordNet:** `nltk.download('wordnet')` dan `nltk.download('omw-1.4')` memastikan data WordNet tersedia untuk NLTK.
2.  **Impor `wordnet`:** Kita mengimpor modul `wordnet` dari `nltk.corpus`.
3.  **`wordnet.synsets(word, lang='eng')`:** Fungsi ini mencari *synset* (set sinonim) untuk kata yang diberikan. Penting untuk dicatat bahwa WordNet yang digunakan NLTK ini adalah WordNet berbahasa Inggris. Jika Anda ingin mencari sinonim untuk kata non-Inggris, Anda harus terlebih dahulu menerjemahkannya ke bahasa Inggris, atau menggunakan sumber daya leksikal lain yang mendukung bahasa tersebut.
4.  **`syn.lemmas(lang='eng')`:** Dari setiap *synset* yang ditemukan, kita mengambil semua *lemma* (bentuk dasar kata) yang merupakan sinonim dalam bahasa Inggris.
5.  **Pembersihan Duplikasi:** Hasilnya kemudian dimasukkan ke dalam `set` untuk menghapus duplikasi dan memberikan daftar sinonim yang unik.

Untuk bahasa Indonesia, WordNet dalam NLTK tidak memiliki cakupan yang kuat. Anda mungkin perlu mempertimbangkan pustaka atau kamus sinonim khusus bahasa Indonesia, atau menggunakan pendekatan berbasis *embedding* jika ingin mencari kata dengan makna serupa.

## Simulasi NLP: Text Classification dengan Scikit-learn dan NLTK

Untuk melakukan klasifikasi teks, kita akan mengikuti langkah-langkah berikut:
1.  **Persiapan Data**: Membuat dataset teks sederhana dengan label.
2.  **Pra-pemrosesan Teks**: Membersihkan dan menormalisasi teks.
3.  **Ekstraksi Fitur**: Mengubah teks menjadi representasi numerik yang dapat dipahami oleh model.
4.  **Pelatihan Model**: Melatih model klasifikasi menggunakan Scikit-learn.
5.  **Evaluasi Model**: Mengukur kinerja model.

```python
import pandas as pd

# Data teks contoh
data = {
    'text': [
        "Natural Language Processing adalah bidang ilmu komputer.",
        "Ilmu komputer dan bahasa manusia adalah interaksi yang menarik.",
        "Manusia berinteraksi dengan komputer melalui bahasa alami.",
        "Cuaca hari ini cerah dengan sedikit awan.",
        "Harga kebutuhan pokok cenderung naik di pasaran.",
        "Berita terbaru mengenai kebijakan pemerintah."
    ],
    'category': [
        'teknologi',
        'teknologi',
        'teknologi',
        'umum',
        'umum',
        'umum'
    ]
}

df_classification = pd.DataFrame(data)
print("Dataset untuk Klasifikasi Teks:")
display(df_classification.head(6))
```

### 2. Pra-pemrosesan Teks

Langkah ini bertujuan untuk membersihkan dan menormalisasi teks agar lebih mudah diolah oleh model. Kita akan menerapkan teknik yang sudah kita pelajari sebelumnya: *lowercasing*, penghapusan tanda baca, dan penghapusan *stop words*.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
import pandas as pd # Import pandas for DataFrame



# Pastikan NLTK data sudah diunduh
nltk.download('punkt') # Memastikan 'punkt' diunduh
nltk.download('stopwords') # Memastikan 'stopwords' diunduh
nltk.download('punkt_tab') # Memastikan 'punkt_tab' diunduh untuk tokenisasi yang tepat

# Dapatkan daftar stop words bahasa Indonesia
stop_words_id = set(stopwords.words('indonesian'))

def preprocess_text(text):
    # 1. Lowercasing
    text = text.lower()

    # 2. Hapus tanda baca dan karakter non-alfanumerik
    text = re.sub(r'[^a-z\s]', '', text) # Hanya menyisakan huruf kecil dan spasi

    # 3. Tokenisasi
    tokens = word_tokenize(text)

    # 4. Hapus stop words
    filtered_tokens = [word for word in tokens if word not in stop_words_id and len(word) > 1]

    # Gabungkan kembali menjadi string
    return ' '.join(filtered_tokens)

# Aplikasikan fungsi preprocessing ke kolom 'text'
df_classification['cleaned_text'] = df_classification['text'].apply(preprocess_text)

print("Dataset setelah pra-pemrosesan:")
display(df_classification.head(6))
```

### 3. Ekstraksi Fitur

Model *machine learning* tidak dapat bekerja langsung dengan teks. Oleh karena itu, kita perlu mengubah teks menjadi representasi numerik. Salah satu metode yang umum digunakan adalah **TF-IDF (Term Frequency-Inverse Document Frequency)**, yang memberikan bobot pada kata berdasarkan seberapa sering muncul dalam dokumen dan seberapa jarang muncul di seluruh korpus dokumen.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Pisahkan fitur (X) dan label (y)
X = df_classification['cleaned_text']
y = df_classification['category']

# Bagi data menjadi training dan testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Jumlah data training: {len(X_train)}")
print(f"Jumlah data testing: {len(X_test)}")

# Inisialisasi TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000) # Batasi jumlah fitur untuk kesederhanaan

# Pelajari kosakata dan ubah data training menjadi vektor TF-IDF
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)

# Ubah data testing menggunakan kosakata yang sama yang dipelajari dari training data
X_test_tfidf = tfidf_vectorizer.transform(X_test)

print("Bentuk X_train_tfidf:", X_train_tfidf.shape)
print("Bentuk X_test_tfidf:", X_test_tfidf.shape)

# Tampilkan beberapa fitur (kata) yang dihasilkan oleh TF-IDF
print("\nContoh nama fitur (kata):", tfidf_vectorizer.get_feature_names_out()[:10])
```

### 4. Pelatihan Model

Setelah teks diubah menjadi fitur numerik, kita dapat melatih model klasifikasi. Kita akan menggunakan **Support Vector Machine (SVM)**, salah satu algoritma yang efektif untuk klasifikasi teks.

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Inisialisasi model SVM
svm_model = SVC(kernel='linear', random_state=42) # Menggunakan kernel linear untuk klasifikasi teks

# Latih model menggunakan data training
svm_model.fit(X_train_tfidf, y_train)

print("Model SVM telah berhasil dilatih.")
```

### 5. Evaluasi Model

Setelah model dilatih, kita perlu mengevaluasi seberapa baik kinerjanya pada data yang belum pernah dilihat sebelumnya (data testing). Kita akan menggunakan metrik seperti *accuracy*, *precision*, *recall*, dan *f1-score*.

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Inisialisasi model SVM (Ditambahkan untuk memperbaiki NameError)
svm_model = SVC(kernel='linear', random_state=42) # Menggunakan kernel linear untuk klasifikasi teks

# Latih model menggunakan data training (Ditambahkan untuk memperbaiki NameError)
svm_model.fit(X_train_tfidf, y_train)

# Lakukan prediksi pada data testing
y_pred = svm_model.predict(X_test_tfidf)

# Hitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi Model: {accuracy:.2f}")

# Tampilkan classification report
report = classification_report(y_test, y_pred)
print("\nClassification Report:\n", report)
```

Penjelasan Hasil Klasifikasi

*   **Akurasi**: Proporsi prediksi yang benar dari semua prediksi. Akurasi 1.00 menunjukkan bahwa model memprediksi semua sampel dengan benar di data testing.
*   **Precision**: Dari semua yang diprediksi sebagai positif untuk suatu kelas, berapa banyak yang benar-benar positif. (True Positives / (True Positives + False Positives))
*   **Recall**: Dari semua yang sebenarnya positif untuk suatu kelas, berapa banyak yang berhasil diprediksi sebagai positif. (True Positives / (True Positives + False Negatives))
*   **F1-Score**: Rata-rata harmonik dari *precision* dan *recall*. Berguna ketika ada ketidakseimbangan kelas.
*   **Support**: Jumlah sampel sebenarnya untuk setiap kelas di data testing.

Dalam kasus ini, dengan dataset yang sangat kecil, akurasi 1.00 dapat dicapai dengan mudah. Pada dataset yang lebih besar dan kompleks, performa model akan bervariasi.

### Uji Model dengan Kalimat Input Baru

Sekarang, mari kita coba prediksi kategori untuk beberapa kalimat baru menggunakan model yang sudah kita latih.

```python
# Kalimat input baru
new_texts = [
    "Belajar machine learning sangat menarik bagi saya.",
    "Kenaikan harga minyak dunia mempengaruhi inflasi.",
    "Kecerdasan buatan merevolusi banyak industri.",
    "Berita terkini menunjukkan pemilu akan segera dilaksanakan."
]

# Buat DataFrame dari kalimat baru
df_new_texts = pd.DataFrame({'text': new_texts})

# Aplikasikan pra-pemrosesan pada teks baru (menggunakan fungsi yang sama)
df_new_texts['cleaned_text'] = df_new_texts['text'].apply(preprocess_text)

# Ekstraksi fitur dari teks baru (menggunakan TF-IDF Vectorizer yang sudah dilatih)
X_new_tfidf = tfidf_vectorizer.transform(df_new_texts['cleaned_text'])

# Lakukan prediksi menggunakan model SVM
new_predictions = svm_model.predict(X_new_tfidf)

# Tampilkan hasil
df_new_texts['predicted_category'] = new_predictions

print("Prediksi untuk Kalimat Baru:")
display(df_new_texts)
```

### Menyimpan dan Memuat Ulang Model

Setelah model dilatih dan dievaluasi, kita dapat menyimpan model tersebut ke dalam file agar dapat digunakan kembali tanpa perlu melatih ulang. Library `joblib` sering direkomendasikan untuk model Scikit-learn.

```python
import joblib

# Tentukan nama file untuk menyimpan model
model_filename = 'svm_text_classifier_model.joblib'
vectorizer_filename = 'tfidf_vectorizer.joblib'

# Simpan model
joblib.dump(svm_model, model_filename)
joblib.dump(tfidf_vectorizer, vectorizer_filename)

print(f"Model SVM telah disimpan ke '{model_filename}'")
print(f"TF-IDF Vectorizer telah disimpan ke '{vectorizer_filename}'")
```

### Memuat Ulang Model dan Menguji Kembali

Untuk memverifikasi bahwa model telah disimpan dengan benar, kita akan memuat ulang model dan vectorizer, lalu menggunakannya untuk memprediksi salah satu kalimat baru.

```python
# Muat ulang model dan vectorizer
loaded_svm_model = joblib.load(model_filename)
loaded_tfidf_vectorizer = joblib.load(vectorizer_filename)

print("Model dan TF-IDF Vectorizer berhasil dimuat ulang.")

# Uji dengan salah satu kalimat baru yang sudah dibersihkan
sample_text = "Kecerdasan buatan merevolusi banyak industri."
sample_cleaned_text = preprocess_text(sample_text)

# Ubah teks contoh menjadi fitur TF-IDF menggunakan vectorizer yang dimuat
sample_tfidf = loaded_tfidf_vectorizer.transform([sample_cleaned_text])

# Lakukan prediksi menggunakan model yang dimuat
loaded_prediction = loaded_svm_model.predict(sample_tfidf)

print(f"\nKalimat asli: '{sample_text}'")
print(f"Kalimat setelah pra-pemrosesan: '{sample_cleaned_text}'")
print(f"Prediksi dari model yang dimuat ulang: '{loaded_prediction[0]}'"
)
```
