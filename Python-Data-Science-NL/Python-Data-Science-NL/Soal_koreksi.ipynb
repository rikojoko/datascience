```python
import random
import os

# 1. Definisikan Kamus Kata Baku (untuk contoh ini, kita buat sederhana)
# Dalam aplikasi nyata, kamus ini akan jauh lebih komprehensif.
kamus_baku = set([
    "pemrosesan", "bahasa", "alami", "nlp", "adalah", "cabang", "dari", 
    "kecerdasan", "buatan", "ai", "yang", "berfokus", "untuk", "memungkinkan",
    "komputer", "memahami", "menafsirkan", "dan", "memanipulasi", "manusia",
    "melalui", "tahapan", "alur", "kerja", "pipeline", "termasuk", "tokenisasi",
    "normalisasi", "teks", "stemming", "lemmatization", "penandaan", "kelas",
    "kata", "part-of-speech", "tagging", "mesin", "dapat", "memproses",
    "tidak", "terstruktur", "mengekstrak", "wawasan", "bermakna", "saat",
    "ini", "mendukung", "banyak", "aplikasi", "sehari-hari", "seperti",
    "asisten", "virtual", "siri", "google", "assistant", "alat", "penerjemah",
    "fitur", "koreksi", "otomatis", "autocorrect", "analisis", "sentimen",
    "filter", "pesan", "spam", "pada", "email", "terlepas", "tantangan",
    "sarkasme", "ketidakpastian", "makna", "ambiguitas", "model", "pembelajaran",
    "mendalam", "deep", "learning", "modern", "telah", "membuat", "komunikasi",
    "menjadi", "lebih", "lancar", "sebelumnya", "binus", "lembaga", "pendidikan",
    "dibidang", "komputer", "makan", "minum", "belajar", "tidur", "buku", 
    "jalan", "cepat", "bagus", "kualitas", "layanan", "lambat", "kecewa",
    "rusak", "buruk", "biasa", "spesial", "rekomendasi", "beli", "jelek",
    "belajar", "menarik", "kenaikan", "harga", "minyak", "dunia", "mempengaruhi",
    "inflasi", "merevolusi", "industri", "berita", "terkini", "pemilu", 
    "akan", "segera", "dilaksanakan", "pemerintah", "meluncurkan", "program",
    "meningkatkan", "kualitas", "indonesia", "efek", "mendalam", "pola",
    "migrasi", "hewan", "keanekaragaman", "hayati", "seluruh", "dunia"
])

print(f"Jumlah kata dalam kamus baku: {len(kamus_baku)}")

# Assuming 'kamus_baku' is available from previous cells
# If not, it should be defined here or sourced from a larger list.

# Function to introduce a single random misspelling into a word
def introduce_misspelling(word):
    if len(word) <= 1: # Cannot easily misspell very short words
        return word + 's' # Just add 's' for simplicity

    word = list(word) # Convert to list for easy manipulation
    error_type = random.choice(['delete', 'insert', 'substitute', 'transpose'])

    if error_type == 'delete':
        idx = random.randint(0, len(word) - 1)
        del word[idx]
    elif error_type == 'insert':
        idx = random.randint(0, len(word))
        char = random.choice('abcdefghijklmnopqrstuvwxyz')
        word.insert(idx, char)
    elif error_type == 'substitute':
        idx = random.randint(0, len(word) - 1)
        char = random.choice('abcdefghijklmnopqrstuvwxyz')
        word[idx] = char
    elif error_type == 'transpose':
        if len(word) >= 2:
            idx1 = random.randint(0, len(word) - 2)
            idx2 = idx1 + 1
            word[idx1], word[idx2] = word[idx2], word[idx1]

    return ''.join(word)

misspelled_words_list = set()
target_count = 500
attempts = 0
max_attempts_per_word = 10 # Prevent infinite loops for un-misspellable words

print(f"Generating {target_count} misspelled words...")

while len(misspelled_words_list) < target_count and attempts < target_count * max_attempts_per_word:
    word_to_misspell = random.choice(list(kamus_baku))
    original_word = word_to_misspell.lower() # Ensure consistent casing

    misspelled = introduce_misspelling(original_word)
    
    # Ensure the generated word is actually misspelled and not already in the dictionary
    if misspelled != original_word and misspelled not in kamus_baku:
        misspelled_words_list.add(misspelled)
    attempts += 1

if len(misspelled_words_list) < target_count:
    print(f"Warning: Could only generate {len(misspelled_words_list)} unique misspelled words after {attempts} attempts.")
else:
    print(f"Successfully generated {len(misspelled_words_list)} unique misspelled words.")

# Convert set to list and shuffle for variety
misspelled_words_output = list(misspelled_words_list)
random.shuffle(misspelled_words_output)

# Save to file
output_filename = "katarusak_words.txt"
with open(output_filename, 'w', encoding='utf-8') as f:
    for word in misspelled_words_output:
        f.write(word + '\n')

print(f"File '{output_filename}' created successfully with {len(misspelled_words_output)} misspelled words.")
```

```python
import os
import string
import difflib

# Path ke file kata salah eja yang sudah dibuat
# Ganti ke 'misspelled_words.txt' karena file ini sudah dibuat sebelumnya
misspelled_file_path = "katarusak_words.txt"

# Baca kata-kata salah eja dari file
misspelled_words_from_file = []
if os.path.exists(misspelled_file_path):
    with open(misspelled_file_path, 'r', encoding='utf-8') as f:
        for line in f:
            word = line.strip()
            if word:
                misspelled_words_from_file.append(word)
    print(f"Berhasil membaca {len(misspelled_words_from_file)} kata salah eja dari '{misspelled_file_path}'.")
else:
    print(f"Error: File '{misspelled_file_path}' tidak ditemukan.")

# print(misspelled_words_from_file) # Komentar ini karena output terlalu panjang

def perbaiki_ejaan_kalimat(kalimat_input):
    # Tahap 1: Normalisasi & Tokenisasi sederhana
    # Mengubah ke huruf kecil dan menghapus tanda baca agar fokus pada kata
    kalimat_clean = kalimat_input.lower().translate(str.maketrans('', '', string.punctuation))
    kata_list = kalimat_clean.split()

    kata_terkoreksi = []
    daftar_typo = []

    # Tahap 2: Deteksi dan Koreksi per Kata
    for kata in kata_list:
        # Jika kata ada di kamus, artinya ejaannya sudah benar
        if kata in kamus_baku:
            kata_terkoreksi.append(kata)
        else:
            # Cari kata paling mirip di kamus menggunakan Levenshtein/Edit Distance
            pilihan_terdekat = difflib.get_close_matches(kata, list(kamus_baku), n=1, cutoff=0.5)

            if pilihan_terdekat:
                kata_terkoreksi.append(pilihan_terdekat[0])
                daftar_typo.append(f"'{kata}' -> '{pilihan_terdekat[0]}'")
            else:
                # Jika terlalu jauh berbeda dan tidak ada yang mirip, biarkan kata aslinya
                kata_terkoreksi.append(kata)

    # Menyatukan kembali token-token kata menjadi kalimat utuh
    kalimat_hasil = " ".join(kata_terkoreksi)

    # Menampilkan Hasil Analysis
    print(f"Kalimat Asli    : {kalimat_input}")
    print(f"Deteksi Typo    : {', '.join(daftar_typo) if daftar_typo else 'Tidak ada typo'}")
    print(f"Hasil Perbaikan : {kalimat_hasil}")
    print("-" * 50)

# --- Uji Coba Program --- 
# Karena misspelled_words_from_file adalah list of words, kita perlu menggabungkannya menjadi satu string (kalimat)
# agar sesuai dengan input yang diharapkan oleh fungsi perbaiki_ejaan_kalimat.
if misspelled_words_from_file:
    sample_sentence = " ".join(misspelled_words_from_file[:20]) # Ambil beberapa kata pertama sebagai contoh kalimat
    print("\n--- Menguji fungsi perbaiki_ejaan_kalimat dengan contoh kalimat ---")
    perbaiki_ejaan_kalimat(sample_sentence)
else:
    print("Tidak ada kata salah eja yang bisa digabungkan menjadi kalimat untuk diuji.")
```
