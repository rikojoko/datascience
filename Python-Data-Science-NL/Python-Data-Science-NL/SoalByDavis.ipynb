````python
from deep_translator import GoogleTranslator
import os
def terjemahkan_file_txt(nama_file_input, nama_file_output):
    # 1. Memeriksa apakah file input yang ingin diterjemahkan ada
    if not os.path.exists(nama_file_input):
        print(f"Error: File '{nama_file_input}' tidak ditemukan.")
        return
    print(f"Membaca file: {nama_file_input}...")
    try:
    # 2. Membaca isi teks dari file .txt asli
        with open(nama_file_input, 'r', encoding='utf-8') as file:
            teks_asli = file.read()
            print(f"Berhasil membaca {len(teks_asli)} karakter dari file.")
            print("-" * 40)
            print("Teks asli:")
            print(teks_asli[:300] + "..." if len(teks_asli) > 300 else teks_asli) 
            print("-" * 40)
        # Jika file ternyata kosong
        if not teks_asli.strip():
            print("Peringatan: File tersebut kosong.")
            return

        print("Sedang memproses penerjemahan dengan NLP...")
        # 3. Menerjemahkan teks (dari Indonesia 'id' ke Inggris 'en')
        # GoogleTranslator otomatis menangani pemotongan teks jika terlalu panjang
        # proses translate ... 

        translator = GoogleTranslator(source='id', target='en')
        teks_terjemahan = translator.translate(teks_asli)

        # 4. Menyimpan hasil terjemahan ke file .txt baru
        with open(nama_file_output, 'w', encoding='utf-8') as file_baru:
            file_baru.write(teks_terjemahan)

        print(f"Sukses! Hasil terjemahan telah disimpan di: '{nama_file_output}'")
        print("\n--- Cuplikan Hasil Terjemahan ---")
        print(teks_terjemahan[:300] + "..." if len(teks_terjemahan) > 300 else teks_terjemahan)

    except Exception as e:
        print(f"Terjadi kesalahan saat memproses file: {e}")
    # --- Cara Penggunaan ---
    # Catatan: Pastikan Anda sudah membuat file bernama 'artikel_indo.txt' di folder yang sama terlebih dahulu

file_input = "artikel_indo.txt"
file_output = "artikel_inggris.txt"

terjemahkan_file_txt(file_input, file_output)
````

## 2. tokenisasi isi file artikel_indo.txt


````python
import nltk
from nltk.tokenize import word_tokenize

with open(file_input, 'r', encoding='utf-8') as file:
    teks_asli = file.read()

tokens = word_tokenize(teks_asli.lower())

print("\n--- Cuplikan Token Hasil Tokenisasi ---")
print(tokens)
````

````python
import nltk
import os
from nltk.tokenize import word_tokenize

file_inggris = "artikel_inggris.txt"

# 1. Memeriksa apakah file artikel_inggris.txt ada
if not os.path.exists(file_inggris):
    print(f"Error: File '{file_inggris}' tidak ditemukan. Pastikan proses translasi sudah dijalankan.")
else:
    # 2. Membaca isi teks dari file artikel_inggris.txt
    with open(file_inggris, 'r', encoding='utf-8') as file:
        teks_inggris = file.read()
    
    # 3. Melakukan tokenisasi setelah mengubah teks menjadi huruf kecil (lowercasing)
    # Parameter language='english' diset agar tokenisasi disesuaikan dengan struktur bahasa Inggris
    tokens_en = word_tokenize(teks_inggris.lower(), language='english')
    
    print(f"Berhasil membaca file: {file_inggris}")
    print(f"Total token bahasa Inggris yang dihasilkan: {len(tokens_en)}")
    print("\n--- Cuplikan Hasil Tokenisasi (Bahasa Inggris) ---")
    print(tokens_en[:20])  # Menampilkan 20 token pertama
````

## 3. Stop Word Removal isi file artikel_indo.txt

````python
from nltk.corpus import stopwords

stop_words_id = set(stopwords.words('indonesian'))

filtered_tokens = [word for word in tokens if word not in stop_words_id and word.isalnum()]

print("\n--- Cuplikan Kata Setelah Stop Word Removal ---")
print(filtered_tokens)
````

## 4. Word Cloud dari Kata-kata Setelah Stop Word Removal

````python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

teks_untuk_wordcloud = " ".join(filtered_tokens)

wordcloud = WordCloud(width=800, height=400, background_color='white', collocations=False).generate(teks_untuk_wordcloud)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud - Artikel Indonesia (Setelah Stop Word Removal)", fontsize=14, pad=20)
plt.show()
````

![word cloud](w_nlp.png)

````python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]

stemmed_tokens = [word for word in stemmed_tokens if word]

print("\n--- Perbandingan Kata Sebelum vs Sesudah Stemming ---")
for asal, hasil in zip(filtered_tokens, stemmed_tokens):
    if asal != hasil:
        print(f"'{asal}' -> menjadi -> '{hasil}'")
    else:
        print(f"'{asal}' -> tetap -> '{hasil}'")
````

````python
import nltk
from nltk.tokenize import word_tokenize

file_input_en = "artikel_inggris.txt"

with open(file_input_en, 'r', encoding='utf-8') as file:
    teks_inggris = file.read()

tokens_en = word_tokenize(teks_inggris.lower())

print("\n--- Cuplikan Token Inggris ---")
print(tokens_en)
````

````python
from nltk.corpus import stopwords

stop_words_en = set(stopwords.words('english'))

filtered_tokens_en = [word for word in tokens_en if word not in stop_words_en and word.isalnum()]

print("\n--- Cuplikan Kata Bersih (Inggris) ---")
print(filtered_tokens_en)
````

````python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

teks_untuk_wordcloud_en = " ".join(filtered_tokens_en)

wordcloud_en = WordCloud(width=800, height=400, background_color='white', collocations=False).generate(teks_untuk_wordcloud_en)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_en, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud - English Article (After Stop Word Removal)", fontsize=14, pad=20)
plt.show()
````

````python
from nltk.stem import PorterStemmer

stemmer_en = PorterStemmer()

stemmed_tokens_en = [stemmer_en.stem(word) for word in filtered_tokens_en]

print("\n--- Perbandingan Sebelum vs Sesudah Stemming (Inggris) ---")
for asal, hasil in zip(filtered_tokens_en, stemmed_tokens_en):
    print(f"'{asal}' -> '{hasil}'")
````
