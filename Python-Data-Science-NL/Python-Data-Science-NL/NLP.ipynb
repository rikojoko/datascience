# Natural Language Processing (NLP) Fundamentals
Natural Language Processing (NLP) adalah bidang ilmu komputer yang berfokus pada interaksi antara komputer dan bahasa manusia. Tujuan utamanya adalah untuk memungkinkan komputer memahami, memproses, dan menghasilkan bahasa manusia.

## Mengapa NLP Penting?
Interaksi Manusia-Komputer: Memungkinkan komputer memahami perintah dan pertanyaan dalam bahasa alami, seperti asisten suara (Siri, Google Assistant).

Analisis Data Tekstual: Mengubah teks mentah menjadi informasi yang terstruktur dan bermakna untuk analisis bisnis, riset pasar, dll.

Otomatisasi: Mengotomatiskan tugas-tugas yang melibatkan bahasa, seperti penerjemahan, ringkasan dokumen, atau penjawab pertanyaan.

## Konsep Dasar NLP
1. Tokenisasi (Tokenization): Proses memecah teks menjadi unit-unit yang lebih kecil (kata, frasa, simbol) yang disebut token.

2. Normalisasi Teks (Text Normalization): Proses membersihkan dan menstandarisasi teks, meliputi:

* Lowercasing: Mengubah semua teks menjadi huruf kecil.

* Stop Word Removal: Menghapus kata-kata umum yang tidak memberikan banyak makna (misalnya, 'dan', 'yang', 'ini').

* Stemming/Lemmatization: Mengurangi kata menjadi bentuk dasarnya (misalnya, 'berlari', 'berlari', 'berlari' -> 'lari').

3. Part-of-Speech Tagging (POS Tagging): Memberi label pada setiap kata dalam teks dengan kategori gramatikalnya (misalnya, kata benda, kata kerja, kata sifat).

4. Named Entity Recognition (NER): Mengidentifikasi dan mengklasifikasikan entitas bernama dalam teks ke dalam kategori yang telah ditentukan (misalnya, nama orang, organisasi, lokasi).

5. Analisis Sentimen (Sentiment Analysis): Menentukan nada emosional di balik sepotong teks (positif, negatif, netral).

## Aplikasi NLP
1. Mesin Penerjemah: Google Translate.

2. Pencarian Informasi: Google Search, Bing.

3. Asisten Virtual: Siri, Google Assistant, Alexa.

4. Analisis Ulasan Produk: Mengidentifikasi sentimen pelanggan.

5. Spam Filtering: Mengklasifikasikan email sebagai spam atau bukan.

## Tokenisasi (Tokenization): 
Proses memecah teks menjadi unit-unit yang lebih kecil (kata, frasa, simbol) yang disebut token.

````python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
````

````python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Teks contoh
text = "Binus merupakan lembaga pendidikan dibidang komputer "

print(f"Teks asli: {text}\n")

# 1. Tokenisasi
tokens = word_tokenize(text.lower()) # Ubah ke huruf kecil sebelum tokenisasi
print(f"Token setelah tokenisasi: {tokens}\n")

# 2. Stop Word Removal
# Dapatkan daftar stop words untuk bahasa Indonesia
stop_words_id = set(stopwords.words('indonesian'))

# Hapus stop words dari token
filtered_tokens = [word for word in tokens if word not in stop_words_id]
print(f"Token setelah penghapusan stop words: {filtered_tokens}")
````

### Hitung frekuensi kemunculan setiap kata

## Visualisasi Frekuensi Kata: Word Cloud

**Word Cloud** adalah representasi visual dari frekuensi kata dalam sebuah teks. Semakin sering sebuah kata muncul, semakin besar ukuran font kata tersebut dalam *word cloud*. Ini adalah cara yang cepat dan menarik untuk mengidentifikasi kata-kata kunci atau topik utama dalam suatu kumpulan teks.

````python
# Hapus tanda baca dari filtered_tokens
cleaned_tokens = [word for word in filtered_tokens if word.isalpha()]

from nltk.probability import FreqDist

# Hitung frekuensi kemunculan setiap kata dari cleaned_tokens
word_frequencies_cleaned = FreqDist(cleaned_tokens)

print("Frekuensi kata setelah penghapusan tanda baca:")
# Tampilkan 10 kata yang paling sering muncul
for word, frequency in word_frequencies_cleaned.most_common(10):
    print(f"{word}: {frequency}")

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Pastikan word_frequencies_cleaned adalah dictionary atau sejenisnya
# WordCloud membutuhkan dictionary {word: frequency}

# Buat objek WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_frequencies_cleaned)

# Tampilkan Word Cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off') # Sembunyikan sumbu x dan y
plt.title('Word Cloud dari Kata-kata Terbersih')

# Simpan wordcloud ke file gambar
plt.savefig('wordcloud_nlp.png', bbox_inches='tight')

plt.show()
````

## Normalisasi Teks (Text Normalization):
Proses membersihkan dan menstandarisasi teks, meliputi:

* Lowercasing: Mengubah semua teks menjadi huruf kecil.

* Stop Word Removal: Menghapus kata-kata umum yang tidak memberikan banyak makna (misalnya, 'dan', 'yang', 'ini').


````python
import re
import string

# 1. Kalimat sampel yang berantakan (banyak huruf kapital, angka, dan tanda baca)
teks_asli = "Wah!! Hari ini tanggal 12 Juli 2026, saya BELAJAR nlp bersama Gemini."

# 2. Mengubah teks menjadi huruf kecil semua (Lowercasing)
teks_lower = teks_asli.lower()

# 3. Menghapus angka menggunakan Regex
teks_tanpa_angka = re.sub(r'\d+', '', teks_lower)

# 4. Menghapus tanda baca (Punctuation) menggunakan string.punctuation
# string.punctuation berisi daftar tanda baca seperti: !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~
teks_bersih = teks_tanpa_angka.translate(str.maketrans('', '', string.punctuation))

# 5. Menghapus spasi berlebih jika ada
teks_normalis = " ".join(teks_bersih.split())

# Menampilkan hasil proses
print("--- Teks Asli ---")
print(teks_asli)
print("\n--- Hasil Tahapan Normalisasi ---")
print(f"1. Huruf Kecil   : {teks_lower}")
print(f"2. Tanpa Angka   : {teks_tanpa_angka}")
print(f"3. Hasil Akhir   : {teks_normalis}")
````

## Simulasi Sederhana NLP: Lematisasi (Lemmatization)

**Lematisasi** adalah proses mengurangi kata ke bentuk dasarnya (lemma) berdasarkan analisis morfologis dan kamus. Berbeda dengan *stemming* yang hanya memotong imbuhan, lematisasi berusaha mengembalikan kata ke bentuk kamus yang benar, sehingga hasilnya seringkali lebih akurat.

````python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
````

````python
from nltk.stem import WordNetLemmatizer

# Inisialisasi lemmatizer
lemmatizer = WordNetLemmatizer()

# Lakukan lemmatisasi pada cleaned_tokens
lemmas = [lemmatizer.lemmatize(word) for word in cleaned_tokens]

print(f"Token setelah penghapusan tanda baca: {cleaned_tokens}")
print(f"Token setelah lematisasi: {lemmas}")
````

 **Stemming**: Proses pemotongan imbuhan (awal/akhir) pada kata secara kasar menggunakan aturan tertentu (rule-based). Tujuannya hanya untuk mengambil kata dasar (disebut stem), namun hasilnya terkadang menghasilkan kata yang tidak baku atau tidak ada di kamus.

````python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# 1. Inisialisasi Stemmer Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# 2. Kalimat sampel dengan kata berimbuhan
teks = "Mereka sedang memakan makanan yang dimasak oleh ibu."

# 3. Proses Stemming
teks_stemmed = stemmer.stem(teks)

print("--- Hasil Stemming (Sastrawi) ---")
print(f"Teks Asli  : {teks}")
print(f"Teks Stem  : {teks_stemmed}")
````

## Menggunakan TextBlob untuk mendeteksi dan membetulkan kata

````python
from textblob import TextBlob

def benahi_ejaan(teks_salah):
    # Menggunakan TextBlob untuk mendeteksi dan membetulkan kata
    blob = TextBlob(teks_salah)
    teks_benar = blob.correct()
    
    print(f"Teks Asli (Typo) : {teks_salah}")
    print(f"Hasil Perbaikan  : {teks_benar}")
    print("-" * 40)

# Uji coba kata serapan bahasa Inggris atau kata yang umum diuji textblob
benahi_ejaan("blak")
benahi_ejaan("natural languge")
````

````python

benahi_ejaan("hme")
````

## "Kamus Kata Baku" sebagai referensi komputer

````python
import difflib

# 1. Kita siapkan "Kamus Kata Baku" sebagai referensi komputer
kamus_baku = ["makan", "minum", "belajar", "komputer", "bahasa", "alami"]

def koreksi_ejaan_lokal(kata_input):
    # Mencari kata di kamus yang paling mirip dengan tingkat kemiripan tertinggi
    pilihan_terdekat = difflib.get_close_matches(kata_input, kamus_baku, n=1, cutoff=0.6)
    
    print(f"Kata diinput    : {kata_input}")
    if pilihan_terdekat:
        print(f"Rekomendasi Baku: {pilihan_terdekat[0]}")
    else:
        print("Rekomendasi Baku: (Tidak ditemukan kata yang mirip di kamus)")
    print("-" * 40)

# Uji coba pembetulan kata berdasarkan kamus_baku di atas
koreksi_ejaan_lokal("makann")   # Kelebihan huruf 'n'
koreksi_ejaan_lokal("bahassa")   # Kurang huruf 'e'
koreksi_ejaan_lokal("komputr")  # Kurang huruf 'e'
````
